```python
# Import necessary packages
import numpy.random as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def build_population(size):
    """
    Build a population of the given size.
    """
    return np.rand(size)
```

```python
def bootstrap_sample(sample_size, size):
    """
    Take a sample from a population and return its mean.
    """
    population = build_population(size)
    sample = np.choice(population, sample_size)
    sample_mean = sample.mean()
    return sample_mean
```

```python
def plot_sample_means(number_samples, sample_size, size):
    """
    Plot the distribution of sample means.
    """
    sample_means = []
    for i in range(0, number_samples):
        sample_means.append(bootstrap_sample(sample_size, size))
    
    label = str(sample_size)
    fig = plt.figure(figsize=(10, 6))
    sns.distplot(sample_means, bins=int(180/5),hist = True,kde = False)
    fig.legend(labels=label)
    plt.show()
```

```python
sample_sizes = [1, 5, 10, 30, 100, 1000, 10000]
for i in sample_sizes:
    plot_sample_means(1000, i, 10000)
```
